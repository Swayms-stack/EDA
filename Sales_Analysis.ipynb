```python
data ['Month'] = data ['Order Date'].dt.month
```

```python
data ['Year'] = data ['Order Date'].dt.year
```

**Q4. Profit Margin for each Sales Record**

```python
data ['Profit Margin'] = data ['Profit'] / data ['Sales']
```

**Q5. Total Profit & Total Sales (by sub-category)**

```python
sub_category = data.groupby('Sub-Category').sum()
```

```python
# Groupby as a function
def grouped_data(column_name):
    '''
    Groupby column and return DataFrame
    Input: Column Name
    '''
    df_tmp = data.groupby(column_name).sum()
    df_tmp.reset_index(inplace=True)
    return df_tmp
    
# Group DataFrame by Segment
grouped_data('Segment')
```

```python
import plotly
import plotly.express as px

fig = px.bar(data,
             x='Sub-Category',
             y='Sales',
             title='<b>Sales by Sub Category</b>',
             template = 'seaborn')
fig.show()
```

**VISUALIZATION FOR PROFITS COLUMN**

```python
fig = px.bar(data,
             x='Sub-Category',
             y='Profit',
             title='<b>Profit by Sub Category</b>',
             template = 'seaborn')
fig.show()
```

*Tables, Supplies & Machines are making losses.*

**Plotting Sales & Profit by Sub-Category**

```python
fig = px.bar(data,
             x='Sub-Category',
             y='Sales', color='Profit',
             color_continuous_scale=["green","red","blue"],
             template = 'seaborn',
             title = '<b>Sales & Profit by Sub Category</b>')
fig.show()
```

```python
fig = px.scatter(data,
                 x='Sales',
                 y='Profit',
                 color='Discount',
                 template = 'seaborn',
                 title = '<b>Scatterplot for Sales & Profit</b>')
fig.show()
```

*Higher Discount resulted in negative profits*

```python
# Aggregating the mean of 'Discount'
discount = data.groupby('Sub-Category').agg({'Discount':'mean',
                                              'Profit':'sum'})
discount
```

**Q6. Mean Discount by Sub Category**

```python
fig = px.bar(discount,
             x=discount.index,
             y='Discount',
             color='Profit',
             color_continuous_scale=['red','green','blue','yellow','pink','purple'],
             template = 'seaborn',
             title = '<b>Mean Discount by Sub Category</b>')

fig.show()
```

**Q7. Plotting Sales & Profit Development for the year 2020**

```python
# Sorting Values by Order Date
sorted = data.sort_values(by=['Order Date'])

# Adding cummulative Sales & Profit
sorted['cummulative_sales'] = sorted['Sales'].cumsum()
sorted['cummulative_profit'] = sorted['Profit'].cumsum()
```

```python
data['Sales'].sum()
```

```python
data['Profit'].sum()
```

```python
fig = px.line(sorted,
              x='Order Date',
              y=['cummulative_sales','cummulative_profit'],
              template = 'seaborn')
              # title= '<b>Sales & Profit Development</b>'
fig.show()
```
